```python
# List of exam scores for 20 students
exam_scores = [75, 85, 90, 95, 60, 45, 80, 72, 50, 38,
              67, 88, 74, 63, 92, 54, 49, 67, 91, 56]

# 1. Overall class performance: Average, Highest, Lowest scores
def overall_performance(scores):
    average_score = sum(scores) / len(scores)
    highest_score = max(scores)
    lowest_score = min(scores)
    return average_score, highest_score, lowest_score

# 2. Pass/Fail classification
def pass_fail_classification(scores, pass_mark=50):
    pass_fail = ["Passed" if score >= pass_mark else "Failed" for score in scores]
    return pass_fail

# 3. Grading on a curve
def assign_grades(scores):
    grades = []
    for score in scores:
        if score >= 90:
            grades.append('A')
        elif score >= 80:
            grades.append('B')
        elif score >= 70:
            grades.append('C')
        elif score >= 50:
            grades.append('D')
        else:
            grades.append('F')
    return grades

# Calculate overall performance
average, highest, lowest = overall_performance(exam_scores)
print(f"Class Average Score: {average:.2f}")
print(f"Highest Score: {highest}")
print(f"Lowest Score: {lowest}")

# Classify students as Passed or Failed
pass_fail_results = pass_fail_classification(exam_scores)
for i, result in enumerate(pass_fail_results, start=1):
    print(f"Student {i}: {result}")

# Assign grades
grades = assign_grades(exam_scores)
for i, grade in enumerate(grades, start=1):
    print(f"Student {i}: Grade {grade}")
```

```python
import cmath

# Function to calculate impedance of a resistor
def impedance_resistor(R):
    return R  # Impedance of resistor is just R (purely real)

# Function to calculate impedance of a capacitor
def impedance_capacitor(C, f):
    omega = 2 * cmath.pi * f  # Angular frequency (omega = 2 * pi * f)
    return 1 / (1j * omega * C)  # Impedance of capacitor (purely imaginary)

# Function to calculate impedance of an inductor
def impedance_inductor(L, f):
    omega = 2 * cmath.pi * f  # Angular frequency (omega = 2 * pi * f)
    return 1j * omega * L  # Impedance of inductor (purely imaginary)

# Function to calculate total impedance in series
def total_impedance_series(impedances):
    return sum(impedances)  # Total impedance in series is the sum of individual impedances

# Function to calculate total impedance in parallel
def total_impedance_parallel(impedances):
    reciprocal_sum = sum(1 / Z for Z in impedances)
    return 1 / reciprocal_sum  # Total impedance in parallel is the reciprocal of the sum of reciprocals

# Example: Impedances of components
R = 10  # Resistance in Ohms
C = 1e-6  # Capacitance in Farads (1 ÂµF)
L = 0.01  # Inductance in Henries (10 mH)
f = 50  # Frequency in Hertz (50 Hz)

# Calculate individual impedances
Z_R = impedance_resistor(R)
Z_C = impedance_capacitor(C, f)
Z_L = impedance_inductor(L, f)

print(f"Impedance of Resistor (R): {Z_R} Ohms")
print(f"Impedance of Capacitor (C): {Z_C} Ohms")
print(f"Impedance of Inductor (L): {Z_L} Ohms")

# Example: Total impedance of components in series
impedances_series = [Z_R, Z_C, Z_L]
Z_total_series = total_impedance_series(impedances_series)
print(f"Total Impedance in Series: {Z_total_series} Ohms")

# Example: Total impedance of components in parallel
impedances_parallel = [Z_R, Z_C, Z_L]
Z_total_parallel = total_impedance_parallel(impedances_parallel)
print(f"Total Impedance in Parallel: {Z_total_parallel} Ohms")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Inventory data
data = {
    'Product': ['Laptop', 'Smartphone', 'Tablet', 'Monitor', 'Keyboard'],
    'Price': [1200, 800, 300, 150, 50],
    'Quantity': [10, 20, 15, 25, 50]
}

# Convert to a DataFrame
df = pd.DataFrame(data)

# 1. Display the inventory
print("Inventory Data:")
print(df)

# 2. Calculate the total inventory value (Price * Quantity for each product)
df['Total_Value'] = df['Price'] * df['Quantity']
total_inventory_value = df['Total_Value'].sum()
print("\nTotal Inventory Value:", total_inventory_value)

# 3. Find the most expensive product
most_expensive_product = df.loc[df['Price'].idxmax()]
print("\nMost Expensive Product:")
print(most_expensive_product[['Product', 'Price']])

# 4. Find the product with the highest stock quantity
highest_stock_product = df.loc[df['Quantity'].idxmax()]
print("\nProduct with the Highest Stock Quantity:")
print(highest_stock_product[['Product', 'Quantity']])

# 5. Calculate the average price of products
average_price = df['Price'].mean()
print("\nAverage Price of Products:", average_price)

# Plotting:

# Plot 1: Bar chart of product quantities
plt.figure(figsize=(10, 5))
plt.bar(df['Product'], df['Quantity'], color='lightblue')
plt.title('Product Quantities')
plt.xlabel('Product')
plt.ylabel('Quantity')
plt.show()

# Plot 2: Pie chart of total inventory value distribution
plt.figure(figsize=(7, 7))
plt.pie(df['Total_Value'], labels=df['Product'], autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title('Total Inventory Value Distribution')
plt.show()
```
